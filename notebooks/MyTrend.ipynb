# Identificando calças

```python
# Carregando a imagem
img_calca = mpimg.imread('/content/img-calca.webp')

# Exibindo a imagem
plt.imshow(img_calca)
plt.axis('off')  # Esconder os eixos
plt.show()
# run inference on a local image

lista = (
    CLIENT.infer(
    "img-calca.webp",
    model_id="mytrend/1"
))

print(lista)

if lista['predictions'][0]['class'] == 'calcas':
  print("****"*15)
  print("___"*15)
  print("")
  print("A imagem contém uma calça!")

  print("___"*15)
  print("")
  print("****"*15)



#if CLIENT.infer
```
